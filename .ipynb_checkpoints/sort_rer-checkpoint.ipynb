```python
data=pd.read_csv('rer_all_audio_files_list_upd.csv')
```

```python
data.drop_duplicates(inplace=True) #remove duplicates in place
data.dropna(subset=['student_id'], inplace=True) #remove NaN subjects
```

```python
duplicates = data[data.duplicated(subset=['student_id', 'task_name', 'item_name', 'semester'], keep=False)]

duplicates=duplicates.sort_values(by=['student_id', 'task_name', 'item_name', 'semester'])
```

```python
#duplicates.to_csv('rer_duplicate_items.csv', index=False)
```

```python
# Get rows of df1 which are not in df2
no_duplicates = data.merge(duplicates, how='outer', indicator=True).query("_merge == 'left_only'").drop('_merge', axis=1)
```

```python
#no_duplicates
#no_duplicates.to_csv('no_duplicates.csv', index=False)
```

```python
# Split the strings in col1 based on ':'
no_duplicates[['task', 'new_col2']] = no_duplicates['task_name'].str.split(':', expand=True)

# Split the strings in new_col2 based on '('
no_duplicates[['task', 'new_col3']] = no_duplicates['task'].str.split('(', expand=True)

# Drop the unwanted columns
no_duplicates.drop(['new_col2', 'new_col3'], axis=1, inplace=True)

# Replace all '_' and ' ' characters in col1 with '' and also remove Practice and Items
no_duplicates['task'] = no_duplicates['task'].str.replace('-', '').str.replace(' ', '').str.replace('Items', '').str.replace('Practice', '')
```

```python
no_duplicates['task'].unique()
```

```python
for index, row in no_duplicates.head(10).iterrows():
    semester = row['semester']
    grade = row['grade']
    audio_string = row['audio_string']
    assessment_name = row['assessment_name']
    task_name = row['task_name']
    item_name = row['item_name']
    student_id = row['student_id']
```

## non word repetition

```python
# Filter rows of df where task is 'NonwordRepetition', 'SentenceRepetition', 'Deletion', or 'Blending'
#new_df = no_duplicates[no_duplicates['task'].isin(['NonwordRepetition', 'SentenceRepetition', 'Deletion', 'Blending'])]
NonwordRepetition_df = no_duplicates[no_duplicates['task'].isin(['NonwordRepetition'])]
```

```python
# Split the strings in col1 based on '_'
NonwordRepetition_df['item'] = NonwordRepetition_df['item_name'].str.split('_', expand=True)[2]
NonwordRepetition_df['assessment_season'] = NonwordRepetition_df['assessment_name'].str.extract('(Fall|Winter|End)')
NonwordRepetition_df['assessment_season'] = NonwordRepetition_df['assessment_season'].str.lower()
NonwordRepetition_df['semester_season'] = NonwordRepetition_df['semester'].str.extract('(fall|winter|spring|end)')
NonwordRepetition_df['true_season'] = NonwordRepetition_df['semester_season'].fillna(NonwordRepetition_df['assessment_season']) #combine semester and assessment season columns
NonwordRepetition_df = NonwordRepetition_df[~NonwordRepetition_df['item'].isin(['S', 'A'])] #remove where item is 'S' or 'A' these don't line up with scores
where_nwr_bad=NonwordRepetition_df['item'].isin(['0013','0009', '0003', '0006', '0027', '0024', '0025', '0018', '0001','0017', '0015', '0016','9003', '9002' ])
NonwordRepetition_df['new_item']= NonwordRepetition_df[where_nwr_bad]['item_name'].str.split('_', expand=True)[3]
NonwordRepetition_df['item'] = NonwordRepetition_df['item'].replace(['0013','0009', '0003', '0006', '0027', '0024', '0025', '0018', '0001','0017', '0015', '0016','9003', '9002'], np.nan)
NonwordRepetition_df['true_item'] = NonwordRepetition_df['item'].fillna(NonwordRepetition_df['new_item']) #combine semester and assessment season columns
#load all_files for matching the folders
all_files=pd.read_csv('all_files.csv',header=None,names=['folder','audio_string'])
# merge the two dataframes on the 'key' column
NonwordRepetition_df = pd.merge(NonwordRepetition_df, all_files, on='audio_string', how='left')
# display the updated dataframe
#merged_df
```
