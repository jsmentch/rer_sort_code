```python
duplicates = data[data.duplicated(subset=['student_id', 'task_name', 'item_name', 'semester'], keep=False)]
```

```python
duplicates
```

```python
duplicates=duplicates.sort_values(by=['student_id', 'task_name', 'item_name', 'semester'])
```

```python
duplicates.to_csv('rer_duplicate_items.csv', index=False)
```

```python
duplicate_ass=duplicates.drop_duplicates(subset=['student_id', 'assessment_name','task_name', 'semester'])
duplicate_ass=duplicate_ass.sort_values(by=['student_id', 'task_name', 'item_name', 'semester'])
duplicate_ass.drop('item_name', axis=1, inplace=True)
duplicate_ass.to_csv('rer_duplicate_assessments.csv', index=False)
```
