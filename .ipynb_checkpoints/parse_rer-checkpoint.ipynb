```python
import os
import random
import shutil
```

```python
# print some random example split filenames from folders by TIME
def print_examples(startdate,enddate,foldir):
    folders = []
    for folder in os.listdir(foldir):
        if folder.isdigit():
            if os.path.isdir(os.path.join(foldir, folder)):
                if startdate <= int(folder) <= enddate:
                    folders.append(folder)

    for folder in folders:
        files = os.listdir(os.path.join(foldir, folder))
        #if files:
        random_selections = random.sample(files, 4)
        
        for filename in random_selections:

            #filename = 'G0_F1_D2-PictureNaming(KF1)-EVO_0_0016_a_U_0-11_18_0_0009-7DAB98.wav'
            split_filename = filename.split('-')
            print(folder)
            print(split_filename)
        print()
# print some better examples
def print_sub_task_item(startdate,enddate,foldir,subind,taskind,itemind,gradeind):
    folders = []
    for folder in os.listdir(foldir):
        if folder.isdigit():
            if os.path.isdir(os.path.join(foldir, folder)):
                if startdate <= int(folder) <= enddate:
                    folders.append(folder)

    for folder in folders:
        print(folder)
        files = os.listdir(os.path.join(foldir, folder))
        #if files:
        random_selections = random.sample(files, 4)
        
        sublist=[]
        tasklist=[]
        itemlist=[]
        gradelist=[]
        for filename in random_selections:

            #filename = 'G0_F1_D2-PictureNaming(KF1)-EVO_0_0016_a_U_0-11_18_0_0009-7DAB98.wav'
            split_filename = filename.split('-')
            
            sublist.append(split_filename[subind])
            task=split_filename[taskind].split('(')[0]
            task=task.split(':')[0]
            task=task.split('Task')[0]
            task=task.split('Practice')[0]
            task=task.split('K')[0]
            task=task.replace('.', '')
            tasklist.append(task)
            itemlist.append(split_filename[itemind])
            gradelist.append(split_filename[gradeind])
            #print(split_filename)
        
        print(f'sub: {sublist}')
        print(f'task: {tasklist}')
        print(f'item: {itemlist}')
        print(f'grade: {gradelist}')
        print()
# parse all of the files in folders based on their start and end dates.
# return lists of files split into sub, task, item, grade, path
def pipe_sub_task_item_grade(startdate,enddate,foldir,subind,taskind,itemind,gradeind):
    folders=[]
    sublist=[]
    tasklist=[]
    itemlist=[]
    gradelist=[]
    folderlist=[]
    filelist=[]
    pathlist=[]
    for folder in os.listdir(foldir):
        if folder.isdigit():
            if os.path.isdir(os.path.join(foldir, folder)):
                if startdate <= int(folder) <= enddate:
                    folders.append(folder)

    for folder in folders:
        print(f'parsing folder: {folder}')
        files = os.listdir(os.path.join(foldir, folder))
        for filename in files:
            filelist.append(filename)
            pathlist.append(os.path.join(foldir, folder, filename))
            folderlist.append(folder)
            #filename = 'G0_F1_D2-PictureNaming(KF1)-EVO_0_0016_a_U_0-11_18_0_0009-7DAB98.wav'
            split_filename = filename.split('-')
            
            sublist.append(split_filename[subind])
            task=split_filename[taskind].split('(')[0]
            task=task.split(':')[0]
            task=task.split('Task')[0]
            task=task.split('Practice')[0]
            task=task.split('K')[0]
            task=task.replace('.', '')
            tasklist.append(task)
            itemlist.append(split_filename[itemind])
            gradelist.append(split_filename[gradeind])
            #print(split_filename)
    print(f'-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-\ncollected {len(filelist)} items\n')
    print(f'unique tasks: {set(tasklist)}\n')
    print(f'unique grades: {set(gradelist)}\n')
    print(f'unique subs: {set(sublist)}\n')
    print(f'unique items: {set(itemlist)}\n')
    print(f'filenames: {filelist[:5]}\n')
    print(f'paths: {pathlist[:5]}\n')
    return(sublist,tasklist,itemlist,gradelist,folderlist,filelist,pathlist,zip(sublist,tasklist,itemlist,gradelist,folderlist,filelist,pathlist))
#         print(f'sub: {sublist}')
#         print(f'task: {tasklist}')
#         print(f'item: {itemlist}')
#         print(f'grade: {gradelist}')
#         print()
```

```python
#sublist,tasklist,itemlist,gradelist,folderlist,filelist,pathlist
rer_path='/om/tmp/jsmentch/rer_test/rer'
readnet_path='/om/tmp/jsmentch/rer_test/readnet'
for i,f in enumerate(filelist):
    sub=sublist[i]
    task=tasklist[i]
    item=itemlist[i]
    grade=gradelist[i]
    folder=folderlist[i]
    path=pathlist[i]
    
    # ReadNet format
    # Task > Semester > Item name > File Name
    # sub-X > ses-X > task > File name
    # ses=grade + season eg K-Fall, G1-Winter, G2-EOY
    truegrade='G0'
    readnet_format_path=os.path.join(readnet_path,f'sub-{sub}',f'ses-{truegrade}_{season}')
    readnet_format=os.path.join(readnet_format_path,f)
    os.makedirs(readnet_format_path, exist_ok=True)
    shutil.copyfile(path, readnet_format)
    
    
    # RER format
    # Task > Semester > Item name > File Name
    if task == 'LetterSounds' or task == 'LetterNaming':
        # Code to be executed if the condition is true
        print("The variable 'task' is either 'LetterSounds' or 'LetterNaming'")
        rer_format_path=os.path.join(rer_path,task,semester)
        rer_format=os.path.join(rer_format_path,f)
        #print(f'RER:\n{rer_format}')
        os.makedirs(rer_format_path, exist_ok=True)
        shutil.copyfile(path, rer_format)
        #print(f'ReadNet:\n{readnet_format}\n')
    else:
        rer_format_path=os.path.join(rer_path,task,semester,item)
        rer_format=os.path.join(rer_format_path,f)
        #print(f'RER:\n{rer_format}')
        os.makedirs(rer_format_path, exist_ok=True)
        shutil.copyfile(path, rer_format)
        #print(f'ReadNet:\n{readnet_format}\n')
```
